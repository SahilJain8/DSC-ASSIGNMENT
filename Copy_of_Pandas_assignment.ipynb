```python
import pandas as pd
```

```python
# Create a series of fruits(string) of length 5 and view it
# Create a series of prices(int) of length 5 and view it
fruits = pd.Series(["Apple", "Banana", "Guava", "Papaya", "Mango"])
prices = pd.Series([150, 25, 100, 30, 75])
print(fruits)
prices
```

```python
# Create a fruit_prices dataframe out of the 2 series you just made and view it
f_p = pd.DataFrame({"Fruits" : fruits, "Price" : prices})
f_p
```

```python
# Find the mean, sum, standard deviation, variance of the prices
print(f_p["Price"].mean())
print(f_p["Price"].sum())
print(f_p["Price"].std())
print(f_p["Price"].var())
```
